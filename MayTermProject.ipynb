```python
#the correlation between age and weight for infants(just writing an easier topic since I do not have the data for the art topic)
#the link for my data: https://www.disabled-world.com/calculators-charts/height-weight-teens.php
#I'm using centimeters since I have no idea about pounds

import matplotlib.pyplot as plt


def runplt():
	plt.figure()
	plt.title("age and weight")
	plt.xlabel("weight-kg")
	plt.ylabel("age-month")
	plt.axis([0,10,0,12])
	plt.grid(True)
	return plt

Y=[[0],[1],[2],[3],[4],[5]]
X=[[3.3],[4.3],[5.3],[6],[6.6],[7.1]]
plt.plot(X,Y,'k.')
plt.show()

from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X,Y)
X2=[[3],[6],[8]]
Y2= model.predict(X2)
plt.plot(X, Y, 'k.')
plt.plot(X2, Y2, 'g-')
plt.show()
```
